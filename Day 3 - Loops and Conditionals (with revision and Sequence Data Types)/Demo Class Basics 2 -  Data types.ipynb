```python
# operations

x = 10
y = 3 

print(x + y) # addition - 13

print(x - y) # subtraction - 7

print(x * y) # multiplication - 30

print(x / y) # division - 3.3333333333333335

print(x // y) # floor division - 3 

print(x % y) # modulus  - 1 - gives you the remainder of the division

print(x ** 2) # exponentiation - 100 # 10 raised to the power 2 # maths.pow(10,2)
```

```python
# libraries
# libraries are pre-written code that we can use to perform specific tasks
# to use a library, we need to import it first

import math 

# we use . to access functions and variables in a library

print(math.sqrt(16)) # square root - 4.0
print(math.pi) # pi constant - 3.141592653589793
print(math.factorial(5)) # factorial - 120
print(math.ceil(4.2)) # ceiling - 5
print(math.floor(4.7)) # floor - 4
print(math.gcd(12, 15)) # greatest common divisor - 3
print(math.lcm(12, 15)) # least common multiple - 60
print(math.log(100, 10)) # logarithm base 10 - 2.0


# https://docs.python.org/3/library/math.html
```

```python
math.sqrt(64)

fact_50 = math.factorial(50)
print(fact_50)

print(type(fact_50)) # <class 'int'>
print(len(str(fact_50))) # 65 - number of digits in 50!

import sys # to access system-specific parameters and functions
print(sys.getsizeof(fact_50)) # size in bytes of the integer
```

```python
# string methods

# methods are functions that are associated with a data types or libraries or object or class

# they are called using dot notation

course = "    python for beginners   "

print(course)
print(course.upper()) # PYTHON FOR BEGINNERS
print(course.lower()) # python for beginners
print(course.title()) # Python For Beginners

print(course.strip()) # removes leading and trailing spaces
print(course.lstrip()) # removes leading spaces
print(course.rstrip()) # removes trailing spaces

print(course.find("for")) # 11 - index of the first occurrence of the substring
print(course.replace("beginners", "absolute beginners"))
print(course.replace(" ", "-")) # replaces all spaces with hyphens

print("python" in course) # True - checks if the substring is present in the string

print(course.count("python"))
print(course.split()) # ['python', 'for', 'beginners'] - splits the string into a list of substrings
```

```python
# boolean 

# boolean value - true or false

# boolean expressions result to either True or False

is_raining = True # variable that holds a boolean value
is_sunny = False # variable that holds a boolean value

print(is_raining)

# python is case-sensitive
is_cloudy = not True
print(is_cloudy) # False

# boolean operations
# boolean operators are or, and, not

# AND
print(True and False) # False

# OR
print(True or False) # True

# NOT
print(not True) # False
print(not False) # True

```

```python
# boolean expressions
# comparison operators

# equal to ==
print(5 == 5) # True

# not equal to !=
print(5 != 3) # True

# greater than >
print(5 > 3) # True

# less than <
print(5 < 3) # False

# greater than or equal to >=
print(5 >= 5) # True

# less than or equal to <=
print(5 <= 3) # False
```
